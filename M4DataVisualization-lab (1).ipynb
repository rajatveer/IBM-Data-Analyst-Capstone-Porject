Connect to the database.

Import pandas module.


```python
import pandas as pd
```


## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)

```

```python
q= """
SELECT name as table_name FROM
sqlite_master WHERE
type='table'
"""

pd.read_sql_query(q, conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

QUERY = """
SELECT * FROM master
"""
df = pd.read_sql_query(QUERY,conn)
df.hist(column='ConvertedComp')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT * FROM master
"""
df = pd.read_sql_query(QUERY,conn)
df.boxplot(column='Age')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY = """
SELECT * FROM master
"""
df = pd.read_sql_query(QUERY,conn)
sns.scatterplot(x= 'Age', y= 'WorkWeekHrs', data=df)
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""

df= pd.read_sql_query(QUERY, conn)

sns.scatterplot(data= df, x='WorkWeekHrs', y='CodeRevHrs', size='Age', hue='Age', alpha=0.5)

plt.title('Bubble plot of Work week hours vs code rev hours')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
q= """
SELECT DatabaseDesireNextYear, COUNT (*) as Count FROM DatabaseDesireNextYear group by DatabaseDesireNextYear order by Count desc limit 5"""

df2= pd.read_sql_query(q, conn)
df2.head()

df2.set_index('DatabaseDesireNextYear', inplace= True)
sizes= df2.iloc[:,0]

df2.plot(kind='pie', 
        figsize= (10,10),
        autopct='%1.1f%%',
        startangle= 90, 
        shadow= True,
        labels= None, 
        pctdistance= 1.12, 
        subplots= True)

plt.title('Top 5 DB Desire for next year')
plt.axis('equal')

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

q= """
SELECT Age, WorkWeekHrs, CodeRevHrs FROM master WHERE Age BETWEEN 30 AND 35"""

df3=pd.read_sql_query(q, conn)
df3
df_plot= df3.groupby('Age').median()
df_plot

df_plot.plot(kind='bar', stacked= True)
plt.title('Stacked Chart')
plt.xlabel('Age')
plt.ylabel("Hours")
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
q="""SELECT Age, ConvertedComp FROM master WHERE Age BETWEEN 45 AND 60"""

df4= pd.read_sql_query(q, conn)
df4

df_lineplot= df4.groupby('Age').median()
df_lineplot


df_lineplot.plot(kind='line')
plt.title('Line Chart')
plt.xlabel('Age')
plt.ylabel('ConvertedComp')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
import numpy as np
q= """
SELECT Respondent, MainBranch FROM master group by MainBranch"""

qwe=pd.read_sql_query(q, conn)
qwe

qwe.plot(kind='barh',figsize= (20,10))
plt.show()
```
